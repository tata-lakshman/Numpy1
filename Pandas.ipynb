             #iloc=>Location based indexing
.iloc is used for integer-location based indexing in pandas DataFrames and Series. It allows you to select data based on its numerical position (from 0 to n-1) in the index, rather than its label. This is in contrast to .loc, which is used for label-based indexing.

For example, s.iloc[3] retrieves the element at the 4th position (index 3) in the Series s, regardless of what the actual index labels are. If you had a DataFrame, you could use .iloc to select rows and columns by their integer position, like df.iloc[0, 1] to get the element in the first row and second column.

```python
print(s.iloc[3])
```

```python
print(s.iloc[[1, 3, 4]])
```

            #loc=>label-based indexing.
Think of your Series s with index labels 'apple', 'banana', 'cherry', etc. When you use s.loc['banana'], pandas looks for the index label 'banana' and returns the corresponding value.

This is different from .iloc, which uses the integer position (0, 1, 2, ...) to select data. So, s.iloc[1] would give you the element at the second position (index 1), which happens to be 'banana' in your current Series, but it's selected by position, not label.

.loc is particularly useful when your index has meaningful labels, as it makes your code more readable and less prone to errors if the order of your data changes.

```python
#loc=>location based indexing
print(s.loc['banana'])
```

#In label-based indexing(loc) your start as well as stop value both are included in the output.

```python
s['banana':'orange']
```

#The main difference between .iloc and .loc in pandas lies in how they select data:

.iloc (integer-location based indexing): This method uses the integer position of the rows and columns to select data. It's like accessing elements in a Python list or array using their numerical index (starting from 0). You use integers or slices of integers with .iloc.
.loc (label-based indexing): This method uses the actual labels of the index and columns to select data. You use the labels (strings, numbers, or a mix) that are present in the index and column names with .loc.

In short:

Use .iloc when you want to select data based on its position.

Use .loc when you want to select data based on its label.

```python
#Dictionary
fruit_proten={
    "Avacado":2.0,
    "Guava":2.6,
    "Blackberry":2.0,
    "Orange":0.9,
    "Banana":1.1,
    "Apple":0.3,
    "Kiwi":1.1,
    "Promegranate":1.7,
    "Mango":0.8,
    "Cherry":1.0
}
s3=pd.Series(fruit_proten,name="Protein")
s3

```

```python
#conditional Selection (true,false)
s3>1
```

s3 > 1: This part creates a boolean Series of the same size as s3. For each element in s3, it checks if the value is greater than 1. If it is, the corresponding value in the boolean Series is True; otherwise, it's False.

s3[...]: When you use this boolean Series inside the square brackets [] after s3, pandas returns only the elements from s3 where the corresponding value in the boolean Series is True.

So, s3[s3 > 1] effectively filters the s3 Series and shows you only the fruits that have a protein value greater than 1.

```python
s3[s3>1]
```
